```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

Pada langkah pertama, saya mengimpor library pandas dan sklearn.preprocessing.OneHotEncoder.
Library pandas digunakan untuk mengelola data dalam bentuk tabel (DataFrame), sedangkan OneHotEncoder dari sklearn berfungsi untuk melakukan proses encoding terhadap kolom kategorikal.

```python
  data = {
    'Warna': ['Merah', 'Biru', 'Hijau', 'Merah', 'Biru'],
    'Ukuran': ['Kecil', 'Sedang', 'Besar', 'Besar', 'Sedang']
}
df = pd.DataFrame(data)
```

\Saya membuat contoh data berupa dua kolom, yaitu Warna dan Ukuran, yang masing-masing berisi data kategorikal seperti “Merah”, “Biru”, “Hijau”, dan “Kecil”, “Sedang”, “Besar”.
Data ini disimpan dalam bentuk DataFrame agar mudah diolah.

```python
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df[['Warna', 'Ukuran']]).toarray()
encoded_df = pd.DataFrame(
    encoded_data,
    columns=encoder.get_feature_names_out(['Warna', 'Ukuran'])
)
```

objek OneHotEncoder() dan menggunakannya untuk melakukan fit_transform pada kolom Warna dan Ukuran.
Hasilnya diubah menjadi array numerik dan kemudian dikonversi menjadi DataFrame baru agar lebih mudah dibaca.

```python
final_df = pd.concat([df, encoded_df], axis=1)
```

Agar hasil lebih mudah dipahami, saya menggabungkan hasil encoding dengan data asli menggunakan pd.concat().

```python
# Import library
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Data contoh
data = {
    'Warna': ['Merah', 'Biru', 'Hijau', 'Merah', 'Biru'],
    'Ukuran': ['Kecil', 'Sedang', 'Besar', 'Besar', 'Sedang']
}

# Membuat DataFrame
df = pd.DataFrame(data)
print("Data Asli:")
print(df)

# Inisialisasi encoder
encoder = OneHotEncoder()

# Melakukan One Hot Encoding
encoded_data = encoder.fit_transform(df[['Warna', 'Ukuran']]).toarray()

# Membuat DataFrame hasil encoding
encoded_df = pd.DataFrame(
    encoded_data,
    columns=encoder.get_feature_names_out(['Warna', 'Ukuran'])
)

print("\nData Setelah One Hot Encoding:")
print(encoded_df)

# Menggabungkan hasil encoding dengan data asli (jika mau)
final_df = pd.concat([df, encoded_df], axis=1)
print("\nData Akhir (Gabungan):")
print(final_df)
```
